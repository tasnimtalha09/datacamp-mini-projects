```python
def validate_user(name, email, password):
    if validate_name(name) == False:
        raise ValueError("Please make sure that your name is greater than 2 characters!")
    if validate_email(email) == False:
        raise ValueError("Your email address is in an incorrect format. Please enter a valid format!")
    if validate_password(password) == False:
        raise ValueError("Your password is too weak. Please ensure that your password includes a capital letter, a number between 0-9 and be greater than 8 characters.")
    return True
```

```python
def register_user(name, email, password):
    if validate_user(name, email, password):
        user = {
            "name" : name,
            "email" : email,
            "password" : password
        }
        return user
    return False
```
