```python
# Read in the Netflix CSV as a DataFrame
pwd = os.getcwd()

netflix_df = pd.read_csv(pwd + "/netflix_data.csv")
netflix_df.head()
```

# Q1: What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called *duration*.

### Creating a Dataset of only **Movies**

```python
# Creating a variable called movie which contains the dataset of only movies
movie = netflix_df[netflix_df["type"] == "Movie"]
movie.head()
```

### Creating a Dataset of movies released in the **1990s**

```python
# Creating a variable called nineties_movies which contains the dataset of only movies released between 1990 and 1999
nineties_movies = movie[(movie["release_year"] >= 1990) & (movie["release_year"] <= 1999)]
nineties_movies.head()
```

### Determining the **mode** of the dataset in **three** ways

```python
# with iloc, we can pinpoint the index number of the rows and columns we want to work with
# With .mode() we can determine the most repeated number in a dataset, which in our case is the most frequent movie duration

duration = nineties_movies.iloc[:, 8].mode()
duration
```

```python
# We can perform the same operation using loc intead of iloc. with loc, we can pinpoint the name of the rows and columns we want to work with

duration_alt = nineties_movies.loc[:, "duration"].mode()
duration_alt
```

```python
# We also can determine the most frequent movie duration by visualising the data in a histogram

plt.hist(nineties_movies["duration"])
plt.title("Distribution of Movies in the 90s")
plt.xlabel("Duration (in minutes)")
plt.ylabel("No. of Movies")
plt.show
```

From the above histogram, we can say that the most frequent movie duration was **arounf 100 minutes** which is pretty close to our actual result of **90 minutes**

## So, the most frequent movie duration is **93 minutes**

# Q2: A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as *short_movie_count*.

## Creating a Dataset of **Short Action** movies

```python
# Creating a variable called action_movies which stores 90s movies with only "action" genre
short_action = nineties_movies[(nineties_movies["genre"] == "Action") & (nineties_movies["duration"] <= 90)]
short_action.head()
```

### Detemining the number of **90s Short Action Movies**

```python
# We can determine the number of rows and columns with the .shape() function

short_movie_count = short_action.shape[0]
short_movie_count
```

## So, there are 9 short action movies that were released in the 90s
